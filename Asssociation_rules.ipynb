```python
# each transcation to list of array items
transcations=data.iloc[:,0].apply(lambda x:x.split(',')).tolist()
transcations
```

```python
# Asssociation rules  Mining
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Make the data into Binary format
# To get the each item in the data per transcation
te = TransactionEncoder()
te_array = te.fit_transform(transcations)
te_array
```

```python
# Convert the array to a DataFrame
basket_df = pd.DataFrame(te_array, columns=te.columns_)
basket_df
```

```python
freq_items=apriori(basket_df,min_support=0.05,use_colnames=True) # minimumn support of 5% 
```

```python
freq_items
```

```python
rules=association_rules(freq_items,metric='lift',min_threshold=1) # min lift ratio of 1
```

```python
rules
```

```python
filtered_rules = rules[(rules['lift']==max(rules['lift']))]# To get the best association rule
```

```python
filtered_rules
```

### Interview Questions

1. What is lift and why is it important in Association rules?

    * Lift measures how much more likely item B is purchased when item A is bought, compared to random chance.

    * It is calculated as:
        Lift(𝐴->𝐵)=Confidence(𝐴->𝐵)/Support(𝐵)

    * Lift > 1 indicates a meaningful and strong relationship between A and B.

2. What is support and confidence? How do you calculate them?

   **Support:** Support shows how frequently an item or itemset occurs:

   -->Support(A) = (Transactions containing A) / (Total transactions)

    **Confidence:** Confidence shows the likelihood of item B being purchased when A is bought:

   -->Confidence(A → B) = (Transactions with A & B) / (Transactions with A)
 
3. What are some limitations or challenges of Association rules mining?
    * It can generate too many rules, many of which may be trivial or irrelevant.
    * Rare but important patterns might be missed due to low support.
    * The technique doesn't consider the sequence or timing of transactions.
    * Works only with categorical or binary (one-hot encoded) data formats.
